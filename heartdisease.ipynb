```python
import pandas as pd

df = pd.read_csv('heart.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# pre-processing

```python
x = df.drop('HeartDisease',axis='columns')
y = df['HeartDisease']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import LabelEncoder
x = x.apply(LabelEncoder().fit_transform)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape , x_test.shape
```

```python
import numpy as np
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
x_train[0]
```

```python
x_test[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train[0]
```

```python
x_test[0]
```

# model definition

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential()

model.add(keras.layers.Input(shape=(11,)))
model.add(keras.layers.Dense(units=128 ,activation='relu'))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(units=128 ,activation='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(units=128 ,activation='relu'))
model.add(keras.layers.Dropout(0.3))
model.add(keras.layers.Dense(units=128 ,activation='relu'))
model.add(keras.layers.Dense(units=1,activation='sigmoid'))

model.compile(optimizer= tf.optimizers.Adam(), loss=tf.losses.binary_crossentropy, metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist = model.fit(x_train,y_train,epochs=200,batch_size=128,validation_data=(x_test,y_test))
```

# plot

```python
import matplotlib.pyplot as plt

plt.plot(hist.history['accuracy'], color = 'blue')
plt.plot(hist.history['val_accuracy'], color = 'red')

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')


```

```python
plt.plot(hist.history['loss'], color = 'blue')
plt.plot(hist.history['val_loss'], color = 'red')

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
```
